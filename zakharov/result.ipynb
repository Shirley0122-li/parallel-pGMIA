```python
import pickle
# package for plotting
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = open('asyn_result', 'rb')
output = pickle.load(file)
[asyn_output_vs_time_mean, asyn_output_vs_time_var, \
    asyn_output_vs_num_mean, asyn_output_vs_num_var] = output
file.close()
```

```python
file = open('syn_result', 'rb')
output = pickle.load(file)
[syn_output_vs_time_mean, syn_output_vs_time_var, \
    syn_output_vs_num_mean, syn_output_vs_num_var] = output
file.close()
```

```python
file = open('sol_result', 'rb')
output = pickle.load(file)
[sol_output_vs_time_mean, sol_output_vs_time_var, \
    sol_output_vs_num_mean, sol_output_vs_num_var] = output
file.close()
```

```python
file = open('pGMIA_result', 'rb')
output = pickle.load(file)
[pGMIA_output_vs_time_mean, pGMIA_output_vs_time_var, \
    pGMIA_output_vs_num_mean, pGMIA_output_vs_num_var] = output
file.close()
```

```python
max_num = 4000
x = np.arange(0,max_num)
fig,ax = plt.subplots()
# ax.plot(asyn_output_vs_num_mean, label = 'parallel')
# ax.plot(seq_output_vs_num_mean, label = 'sequential')

ax.plot(asyn_output_vs_num_mean[0:max_num], label = 'Asynchronous pGMIA', color='dodgerblue')
ax.fill_between(x,asyn_output_vs_num_mean[0:max_num]-2*asyn_output_vs_num_var[0:max_num]/np.sqrt(150), \
    asyn_output_vs_num_mean[0:max_num]+2*asyn_output_vs_num_var[0:max_num]/np.sqrt(150), \
        alpha=0.2, facecolor='dodgerblue')
ax.plot(syn_output_vs_num_mean[0:max_num], label = 'Synchronous pGMIA', linestyle='--', color='darkorange')
ax.fill_between(x,syn_output_vs_num_mean[0:max_num]-2*syn_output_vs_num_var[0:max_num]/np.sqrt(100), \
    syn_output_vs_num_mean[0:max_num]+2*syn_output_vs_num_var[0:max_num]/np.sqrt(100), \
        alpha=0.2, facecolor='darkorange')
ax.plot(sol_output_vs_num_mean[0:max_num], label = 'Batch pGMIA', linestyle='-.', color='forestgreen')
ax.fill_between(x,sol_output_vs_num_mean[0:max_num]-2*sol_output_vs_num_var[0:max_num]/np.sqrt(100), \
    sol_output_vs_num_mean[0:max_num]+2*sol_output_vs_num_var[0:max_num]/np.sqrt(100), \
        alpha=0.2, facecolor='forestgreen')
ax.plot(pGMIA_output_vs_num_mean[0:max_num]-1, label = 'Sequential pGMIA', linestyle=':', color='deeppink')
ax.fill_between(x,pGMIA_output_vs_num_mean[0:max_num]-1-2*pGMIA_output_vs_num_var[0:max_num]/np.sqrt(100), \
    pGMIA_output_vs_num_mean[0:max_num]-1+2*pGMIA_output_vs_num_var[0:max_num]/np.sqrt(100), \
        alpha=0.2, facecolor='deeppink')

ax.set_xlabel('Number of simulations')
ax.set_ylabel('True Optimality Gap')
ax.set_ylim([0,30])
ax.legend()
plt.savefig('zak_num.png')
```
